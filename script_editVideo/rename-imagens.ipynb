```python
import os
import re

pattern = r'\d+'

# Função para extrair o número inteiro do nome do arquivo
def extract_number(image_files):
    return int(re.findall(pattern, image_files)[0])

def renomear_arquivos(pasta, prefixo):
    # Verificar se a pasta existe
    if not os.path.exists(pasta):
        print(f'A pasta "{pasta}" não existe.')
        return

    # Listar arquivos na pasta
    image_files = [f for f in os.listdir(pasta) if f.endswith('.png')]

    # Ordenar arquivos pela numeração
    image_files = sorted(image_files, key=extract_number)

    # Inicializar o contador
    contador = 0

    # Renomear cada arquivo na pasta
    for image_file in image_files:
        # Criar novo nome com o prefixo e contador com zero à esquerda
        novo_nome = f'{prefixo}_{contador:06d}.png'
        # Renomear o arquivo
        os.rename(os.path.join(pasta, image_file), os.path.join(pasta, novo_nome))
        print(f'"{image_file}" renomeado para "{novo_nome}"')
        # Incrementar contador
        contador += 1

if __name__ == '__main__':
    pasta = 'video1-label'
    prefixo = 'video1_img'

    renomear_arquivos(pasta, prefixo)
```
